## Change ``highway MPG`` (MPG = Miles Per Gallon) to ``Highway L/100km`` and ``city mpg`` to ``City L/100km``

### Creating MPG to L/100km function

```python
def mpg_to_lp100km(mpg) :
    # 1 mpg = 235.214583 l/100km
    return 235.214583/mpg

mpg_to_lp100km(50)
```

### Rename the fields

```python
cars_db.update_many( {}, {"$rename":{"highway MPG": "Highway L/100km", "city mpg": "City L/100km"}})
cars_db.find_one()
```

### Convert values to L/100km

```python
cars_db.update_many({}, {"$mul":{"Highway L/100km": 1/235.214583}})
```

```python
cars_db.find_one()
```

### Horsepower per cylinder

```python
cars_db.update_many({"Engine Cylinders" : {"$gt": 0}}, [{"$set":{"Horsepower per cylinder": {"$divide" : ["$Engine HP" , "$Engine Cylinders"]}}}])
```

```python
cars_db.find_one()
```

### Show the top 5 HP/cylinder

```python
[car for car in cars_db.find({}, {"_id": 0, "Make":1, "Model": 1, "Engine HP": 1, "Engine Cylinders": 1,"Horsepower per cylinder":1}).sort("Horsepower per cylinder", -1)][:5]
```
